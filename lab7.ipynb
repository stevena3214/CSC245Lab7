```python
import numpy as np
import pandas as pd 
from sklearn.datasets import load_breast_cancer 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


cancer = load_breast_cancer() 
print(cancer.DESCR) # Print the data set description
```

```python
cancer.data.shape[1]
```

```python
cancerdf = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
cancerdf['target'] = cancer.target
print (cancerdf)
```

```python
cancerdf['target'].value_counts().rename(index={0: 'malignant', 1: 'benign'})
```

```python
def question_3():
    X = cancerdf.drop('target', axis=1)
    y = cancerdf['target']
    return X, y
print(question_3())
```

```python
def question_4():
    X, y = question_3()
    return train_test_split(X, y, test_size=0.25, random_state=0)
print(question_4())
```

```python
def question_5():
    X_train, X_test, y_train, y_test = question_4()
    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train, y_train)
    return knn
print(question_5())
```

```python
def question_6():
    knn = question_5()
    X_mean = pd.DataFrame([cancerdf.mean()[:-1].values], columns=cancer.feature_names)
    return knn.predict(X_mean)
print(question_6())
```

```python
def question_7():
    knn = question_5()
    _, X_test, _, _ = question_4()
    return knn.predict(X_test)
print(question_7())
```

```python
def question_8():
    knn = question_5()
    _, X_test, _, y_test = question_4()
    return knn.score(X_test, y_test)
print(question_8())
```
